# **Important Libraries**

```python
import google.generativeai as genai
from IPython.display import Markdown
from google.colab import userdata
import httpx
import whisper  # for audio inputs
from google.colab import files
```

# **API Configuration**

```python
GOOGLE_API_KEY=userdata.get('F')
genai.configure(api_key=GOOGLE_API_KEY)
```

# **Sample Audio**

```python
uploaded = files.upload()
audio_path = list(uploaded.keys())[0]

model = whisper.load_model("base")

result = model.transcribe(audio_path)
transcribed_text = result["text"]
```

# **Model Response**

```python
gemini_model = genai.GenerativeModel("gemini-1.5-pro")
response = gemini_model.generate_content(f"Analyze this speech:\n\n{transcribed_text}")

print("\n Transcribed Text:\n", transcribed_text)
print("\n Gemini's Analysis:\n", response.text)
```
